```python
from sklearn import datasets
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
tips=sns.load_dataset('tips')
titanic=pd.read_csv('tested.csv')
flights=sns.load_dataset('flights')
iris=sns.load_dataset('iris')
```

```python
tips.sample(5)
```

**1. Scatter plot**(Numerical--Numerical)

```python
sns.scatterplot(tips['total_bill'],tips['tip'],hue=tips['sex'])
```

```python
sns.scatterplot(tips['total_bill'],tips['tip'],hue=tips['sex'],style=tips['smoker'])
```

```python
sns.scatterplot(tips['total_bill'],tips['tip'],hue=tips['sex'],style=tips['sex'],size=tips['size'])
```

**2. Bar plot** (Numrical -- categorical)

```python
titanic.sample(5)
```

```python
sns.barplot(titanic['Pclass'],titanic['Age'])
```

```python
sns.barplot(titanic['Pclass'],titanic['Fare'])
```

```python
sns.barplot(titanic['Pclass'],titanic['Age'],hue=titanic['Sex'])
```

**3. Box Plott**(Numerical--categorical)

```python
sns.boxplot(titanic['Sex'],titanic['Age'],hue=titanic['Survived'])
```

**4. Dist plot** (Numrical -- categorical)

```python
sns.distplot(titanic[titanic['Survived']==0]['Age'],hist=False)
sns.distplot(titanic[titanic['Survived']==1]['Age'],hist=False)
```

**5. HistMap** (Categorical -- categorical)

```python
titanic.head()
```

```python
sns.heatmap(pd.crosstab(titanic['Pclass'],titanic['Survived']))
```

```python
titanic.groupby('Pclass').mean()['Survived']*100
```

**6. ClusterMap** (Numrical -- categorical)

```python
sns.clustermap(pd.crosstab(titanic['Parch'],titanic['Survived']))
```

**7. Pair plot** 

```python
sns.pairplot(iris , hue='species')
```

**8. LinePlot**

```python
flights.sample(5)
```

```python
new=flights.groupby('year').sum().reset_index()
```

```python
sns.lineplot(new['year'],new['passengers'])
```

```python
sns.heatmap(flights.pivot_table(values='passengers',index='month',columns='year'))
```
