**Tokenization**

---
Tokenization is the process of turning a meaningful piece of data, such as an account number, into a random string of characters called a token that has no meaningful value if breached. Tokens serve as reference to the original data, but cannot be used to guess those values.



```python
#importing the NL toolkit(text preprocessing)
import nltk
# downloding all the libraries of nltk
nltk.download('all')
```

```python
#Tokenization of paragraph/sentences
 
p="""Articles by American writer John Neal were published in newspapers, magazines, and literary journals and are part of his bibliography. Neal was the United States' first art critic, author of the first history of American literature, the first American to be published in any British literary magazine, and one of the first male advocates of women's rights and feminist causes in the United States. As an early and outspoken theater critic, he drafted a future for American drama that was only partially realized sixty years later. His critiques of literature helped launch the careers of many well-known American authors and his essays on art were recognized as "prophetic". One of the leading critics of his time, his writing also addressed gender, race, slavery, children, education, law, politics, architecture, religion, gymnastics, civics, American history, science, phrenology, travel, language, political economy, and temperance. Literary historian Fred Lewis Pattee found that "his critical judgments have held. Where he condemned, time has almost without exception condemned also."""

#Convert para into sentence
s=nltk.sent_tokenize(p)
for i in s:
  print(i) 
#Convert para into word
s=nltk.word_tokenize(p)
print(s)
```

**Stemming**

---
Stemming is the process of producing morphological variants of a root/base word. Stemming programs are commonly referred to as stemming algorithms or stemmers. A stemming algorithm reduces the words “chocolates”, “chocolatey”, “choco” to the root word, “chocolate” and “retrieval”, “retrieved”, “retrieves” reduce to the stem “retrieve”.


```python
#import nltk(already imported)
from nltk.stem import PorterStemmer # importing the stemming library for stemming
from nltk.corpus import stopwords #used to remove unwanted and repeated word 


p="""Articles by American writer John Neal were published in newspapers, magazines, and literary journals and are part of his bibliography. Neal was the United States' first art critic, author of the first history of American literature, the first American to be published in any British literary magazine, and one of the first male advocates of women's rights and feminist causes in the United States. As an early and outspoken theater critic, he drafted a future for American drama that was only partially realized sixty years later. His critiques of literature helped launch the careers of many well-known American authors and his essays on art were recognized as "prophetic". One of the leading critics of his time, his writing also addressed gender, race, slavery, children, education, law, politics, architecture, religion, gymnastics, civics, American history, science, phrenology, travel, language, political economy, and temperance. Literary historian Fred Lewis Pattee found that "his critical judgments have held. Where he condemned, time has almost without exception condemned also."""

s=nltk.sent_tokenize(p)#para in sentence
stemmer=PorterStemmer()#Creating the object of Stemmer
#stemming
for i in range(len(s)):
  w=nltk.word_tokenize(s[i])#word tokenize
  w=[stemmer.stem(ws) for ws in w if ws not in set(stopwords.words("english"))] #do stemming to those word who is not present in stopword set
  s[i]=' '.join(w)# coverting list of word into sentence again
  print(s[i])
  print()
```

**Lemmatization**

---
Lemmatization is the process of grouping together the different inflected forms of a word so they can be analyzed as a single item. Lemmatization is similar to stemming but it brings context to the words.

---
Examples of lemmatization:

1. rocks : rock
2. corpora : corpus
3. better : good


```python
#import nltk#importing the NL toolkit(already imported)
from nltk.stem import WordNetLemmatizer

p="""Articles by American writer John Neal were published in newspapers, magazines, and literary journals and are part of his bibliography. Neal was the United States' first art critic, author of the first history of American literature, the first American to be published in any British literary magazine, and one of the first male advocates of women's rights and feminist causes in the United States. As an early and outspoken theater critic, he drafted a future for American drama that was only partially realized sixty years later. His critiques of literature helped launch the careers of many well-known American authors and his essays on art were recognized as "prophetic". One of the leading critics of his time, his writing also addressed gender, race, slavery, children, education, law, politics, architecture, religion, gymnastics, civics, American history, science, phrenology, travel, language, political economy, and temperance. Literary historian Fred Lewis Pattee found that "his critical judgments have held. Where he condemned, time has almost without exception condemned also."""

s=nltk.sent_tokenize(p)#para in sentence
lemma=WordNetLemmatizer()#Creating the object of Lemmatizer
for i in range(len(s)):
  words=nltk.word_tokenize(s[i])
  words=[lemma.lemmatize(word) for word in words if word not in set(stopwords.words('english'))]#do lemmatizing to those word who is not present in stopword set
  s[i]=' '.join(words)
  print(s[i])
  print()
```

**Matcher**

```python
#importing the required module
import spacy
from spacy.matcher import Matcher
from spacy.tokens import Span
import spacy.displacy as displacy
```

```python
#load the english web core
nlp=spacy.load("en_core_web_sm")
```

```python
#performing token based matching
pattern=[{'LOWER':'hello'},{'IS_PUNCT':True,'OP':'?'},{'LOWER':"world"}]
matcher=Matcher(nlp.vocab)#Creating object of matcher
matcher.add('name',None,pattern)
#we create doc once again after adding pattern to the matcher
doc=nlp('Hello,World!')
matches=matcher(doc)
matches
for match_id,start,end in matches:
  string_id=nlp.vocab.strings[match_id]
  span=doc[start:end]
  print(match_id,string_id,start,end,span.text)  
```

```python
#RE in SpaCy
import spacy
import re
doc = nlp("The United States of America (USA) are commonly known as the United States (U.S. or US) or America.")

expression = r"[Uu](nited|\.?) ?[Ss](tates|\.?)"
for match in re.finditer(expression, doc.text):
    start, end = match.span()
    span = doc.char_span(start, end)
    # This is a Span object or None if match doesn't map to valid token sequence
    if span is not None:
        print("Found match:", span.text)
```

**Shallow Parsing**

---
Shallow parsing, or chunking, is the process of extracting phrases from unstructured text. Chunking groups adjacent tokens into phrases on the basis of their POS tags. There are some standard well-known chunks such as noun phrases, verb phrases, and prepositional phrases.



**Noun Phrase Detection**

---


A noun phrase is a phrase that has a noun as its head. It could also include other kinds of words, such as adjectives, ordinals, determiners. Noun phrases are useful for explaining the context of the sentence. They help you infer what is being talked about in the sentence.

```python
conference_text = ('There is a developer conference happening on 21 July 2019 in London.')
conference_doc = nlp(conference_text)
# Extract Noun Phrases
for chunk in conference_doc.noun_chunks:
  print(chunk)

```
